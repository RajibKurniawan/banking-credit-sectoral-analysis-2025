# i. Background

Kredit investasi merupakan salah satu komponen utama dalam aktivitas perbankan yang berperan penting dalam mendorong pertumbuhan ekonomi jangka panjang. Melalui penyaluran kredit investasi, perbankan mendukung ekspansi bisnis, pembangunan infrastruktur, serta penguatan sektor-sektor produktif di berbagai bidang ekonomi.

Namun, penyaluran kredit investasi tidak selalu merata antar sektor ekonomi. Konsentrasi kredit pada sektor tertentu dapat meningkatkan risiko portofolio, terutama ketika terjadi perlambatan ekonomi atau shock sektoral. Oleh karena itu, pemahaman terhadap distribusi, tren, dan dinamika kredit investasi berdasarkan sektor ekonomi menjadi krusial bagi perbankan dalam mengambil keputusan strategis.

Dengan memanfaatkan data posisi kredit investasi perbankan tahun 2025, analisis ini bertujuan untuk memberikan gambaran menyeluruh mengenai pola penyaluran kredit, mengidentifikasi sektor-sektor dominan maupun sektor yang menunjukkan pertumbuhan signifikan, serta menilai potensi risiko dan peluang dari perspektif bisnis dan manajemen risiko.

### Objective
Project ini bertujuan untuk menganalisis posisi dan tren kredit investasi perbankan berdasarkan sektor ekonomi dengan pendekatan data-driven, guna menghasilkan insight yang relevan bagi pengambilan keputusan bisnis.

```python
# jadikan baris 0 sebagai header
data.columns = data.iloc[0]
data = data.drop(index=0).reset_index(drop=True)

# rename kolom sektor
data = data.rename(columns={data.columns[0]: "Sektor Ekonomi (IE)"})
```

```python
bulan = [
    "Januari",
    "Februari",
    "Maret",
    "April",
    "Mei",
    "Juni",
    "Juli",
    "Agustus",
    "September",
    "Oktober",
    "November",
    "Desember",
    "Tahunan",
]
```

```python
data_rupiah = data.iloc[:, 0:14]
data_rupiah.columns = ["Sektor Ekonomi (IE)"] + bulan

data_valas = data.iloc[:, [0] + list(range(14, 27))]
data_valas.columns = ["Sektor Ekonomi (IE)"] + bulan
```

```python
# buang kolom sektor di valas agar tidak dobel
data_valas_only = data_valas.iloc[:, 1:]

# gabungkan kiri (rupiah) dan kanan (valas)
data_clean = pd.concat([data_rupiah, data_valas_only], axis=1)
```

```python
# jumlah kolom
n_rupiah = data_rupiah.shape[1]
n_valas = data_valas_only.shape[1]

# level 0 (tahun & jenis)
level_0 = [""] + ["2025 Rupiah"] * (n_rupiah - 1) + ["2025 Valuta Asing"] * n_valas

# level 1 (nama kolom asli)
level_1 = data.columns.tolist()

# buat multiindex
data.columns = pd.MultiIndex.from_arrays([level_0, level_1])
```

# iv. Cleaning

```python
# Strip whitespace di setiap level MultiIndex
data.columns = pd.MultiIndex.from_tuples(
    [(str(level0).strip(), str(level1).strip()) for level0, level1 in data.columns]
)
```

```python
# Pastikan kolom sektor bertipe string
data.iloc[:, 0] = data.iloc[:, 0].astype(str)

# Drop baris Total / Catatan
data = data[~data.iloc[:, 0].str.contains("Total|Catatan", case=False, na=False)]
```

```python
# Ambil semua kolom kecuali sektor
numeric_cols = data.columns[1:]

for col in numeric_cols:
    data[col] = (
        data[col]
        .astype(str)
        .str.replace(".", "", regex=False)
        .str.replace(",", ".", regex=False)
    )
    data[col] = pd.to_numeric(data[col], errors="coerce")
```

```python
for currency in ["2025 Rupiah", "2025 Valuta Asing"]:
    data[currency] = data[currency].interpolate(axis=1)
```

```python
# Cek missing values
data.isnull().sum()

# Cek nilai negatif
(data[numeric_cols] < 0).sum()
```

```python
def format_rupiah(x):
    if pd.isna(x):
        return "-"
    return "Rp {:,.0f}".format(x).replace(",", ".")
```

```python
data_display = data.copy()

# ambil kolom numerik saja
numeric_cols = data_display.columns[1:]

for col in numeric_cols:
    data_display[col] = data_display[col].apply(format_rupiah)

data_display
```

```python
data.style.format(format_rupiah, subset=numeric_cols)

data.style.format(format_rupiah, subset=numeric_cols).set_properties(
    **{"text-align": "right"}
)
```

# v. Exploratory Data Analysis (EDA)

##### 1. Statistik Deskriptif Kredit Investasi

Tujuan

Melihat gambaran umum:
- skala kredit per sektor
- perbandingan Rupiah vs Valuta Asing

```python
# Statistik deskriptif kredit Rupiah
rupiah_desc = data["2025 Rupiah"].describe()
valas_desc = data["2025 Valuta Asing"].describe()

rupiah_desc, valas_desc
```

Insight :
Kredit investasi dalam Rupiah memiliki nilai rata-rata dan total yang jauh lebih besar dibandingkan Valuta Asing, menunjukkan dominasi pembiayaan domestik dalam portofolio kredit investasi perbankan.

##### 2. Total Kredit Investasi per Sektor (Tahunan)
Tujuan:
- Mengidentifikasi sektor ekonomi dominan.

```python
# Ambil kolom tahunan
data["Total Kredit"] = (
    data["2025 Rupiah"]["Tahunan"] + data["2025 Valuta Asing"]["Tahunan"]
)

# Sort sektor berdasarkan total kredit
sector_total = data[["Total Kredit"]].sort_values(by="Total Kredit", ascending=False)

sector_total.head()
```

```python
# Visualisasi

plt.figure(figsize=(10, 6))
plt.barh(sector_total.index, sector_total["Total Kredit"])
plt.gca().invert_yaxis()
plt.title("Total Kredit Investasi per Sektor Ekonomi (2025)")
plt.xlabel("Nilai Kredit")
plt.ylabel("Sektor Ekonomi")
plt.tight_layout()
plt.show()
```

Insight : 
Kredit investasi terkonsentrasi pada beberapa sektor utama, mengindikasikan potensi risiko konsentrasi portofolio apabila sektor tersebut mengalami tekanan ekonomi.

##### 3. Perbandingan Rupiah vs Valuta Asing per Sektor
Tujuan :
- Melihat exposure mata uang per sektor.

```python
currency_comparison = pd.DataFrame(
    {
        "Rupiah": data["2025 Rupiah"]["Tahunan"],
        "Valuta Asing": data["2025 Valuta Asing"]["Tahunan"],
    }
)

currency_comparison.head()
```

```python
# Visualisasi

currency_comparison.plot(kind="bar", stacked=True, figsize=(10, 6))
plt.title("Komposisi Kredit Investasi Rupiah vs Valuta Asing per Sektor (2025)")
plt.xlabel("Sektor Ekonomi")
plt.ylabel("Nilai Kredit")
plt.legend(title="Jenis Kredit")
plt.tight_layout()
plt.show()
```

Insight : Sebagian besar sektor masih didominasi kredit Rupiah, sementara kredit Valuta Asing lebih terkonsentrasi pada sektor tertentu yang berorientasi ekspor atau memiliki eksposur internasional.

##### 4. Tren Bulanan Kredit Investasi (All Sectors)
Tujuan :
- Menilai dinamika kredit investasi sepanjang tahun.

```python
monthly_cols = bulan[:-1]  # exclude 'Tahunan'

monthly_trend_rupiah = data["2025 Rupiah"][monthly_cols].sum()
monthly_trend_valas = data["2025 Valuta Asing"][monthly_cols].sum()
```

```python
# Visualisasi

plt.figure(figsize=(10, 6))
plt.plot(monthly_trend_rupiah.index, monthly_trend_rupiah.values, label="Rupiah")
plt.plot(monthly_trend_valas.index, monthly_trend_valas.values, label="Valuta Asing")
plt.title("Tren Bulanan Kredit Investasi Perbankan Tahun 2025")
plt.xlabel("Bulan")
plt.ylabel("Total Kredit")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Insight: Kredit investasi menunjukkan tren meningkat pada paruh kedua tahun 2025, yang mengindikasikan peningkatan aktivitas investasi dan ekspansi bisnis.

##### 5. Volatilitas Kredit per Sektor (Risk Perspective)
Tujuan :
- Mengidentifikasi sektor dengan fluktuasi tinggi.

```python
volatility = data["2025 Rupiah"][monthly_cols].std(axis=1)

volatility_df = volatility.sort_values(ascending=False)
volatility_df.head()
```

insight : Sektor dengan volatilitas tinggi memerlukan perhatian khusus dalam pengelolaan risiko kredit karena menunjukkan ketidakstabilan penyaluran kredit sepanjang tahun.

Exploratory Data Analysis menunjukkan bahwa kredit investasi perbankan tahun 2025 masih didominasi oleh kredit dalam Rupiah dengan konsentrasi pada sektor ekonomi tertentu. Tren bulanan menunjukkan peningkatan pada semester kedua, sementara tingkat volatilitas antar sektor bervariasi, mencerminkan perbedaan profil risiko dan karakteristik bisnis masing-masing sektor.

##### 6. Korelasi Kredit Rupiah dan Valuta Asing (Tahunan)
Tujuan:
- Menilai hubungan antara kredit Rupiah dan Valuta Asing per sektor.

```python
# Data korelasi tahunan
corr_data = pd.DataFrame(
    {
        "Rupiah": data["2025 Rupiah"]["Tahunan"],
        "Valuta Asing": data["2025 Valuta Asing"]["Tahunan"],
    }
)

# Hitung korelasi
corr_matrix = corr_data.corr()
corr_matrix
```

```python
# Visualisasi heatmap korelasi
plt.figure(figsize=(5, 4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", center=0)
plt.title("Korelasi Kredit Rupiah dan Valuta Asing (Tahunan)")
plt.tight_layout()
plt.show()
```

Insight:
Terdapat korelasi positif antara kredit Rupiah dan Valuta Asing, yang menunjukkan bahwa sektor dengan penyaluran kredit Rupiah yang besar cenderung juga memiliki eksposur kredit Valuta Asing. Hal ini mengindikasikan bahwa ekspansi kredit lintas mata uang umumnya terjadi pada sektor ekonomi yang sama.


# x. Datasave

```python
# Ambil kolom bulan (tanpa Tahunan)
bulan_cols = bulan[:-1]

# Rupiah (long)
rupiah_long = data["2025 Rupiah"][bulan_cols].copy()
rupiah_long["Sektor Ekonomi"] = data.iloc[:, 0].values

rupiah_long = rupiah_long.melt(
    id_vars="Sektor Ekonomi", var_name="Bulan", value_name="Nilai Kredit"
)
rupiah_long["Jenis Kredit"] = "Rupiah"

# Valas (long)
valas_long = data["2025 Valuta Asing"][bulan_cols].copy()
valas_long["Sektor Ekonomi"] = data.iloc[:, 0].values

valas_long = valas_long.melt(
    id_vars="Sektor Ekonomi", var_name="Bulan", value_name="Nilai Kredit"
)
valas_long["Jenis Kredit"] = "Valuta Asing"

# Gabungkan
tableau_data = pd.concat([rupiah_long, valas_long], ignore_index=True)
```

```python
tableau_data.to_csv("credit_investment_2025_data_clean.csv", index=False)
```
